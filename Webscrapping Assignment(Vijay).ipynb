```python
from bs4 import BeautifulSoup
import requests
import html5lib

url='https://www.imdb.com/list/ls063528769/'

response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.text,'html5lib')
```

```python
print(soup.find('title').text)
```

```python
movies=soup.select('div.lister-list.lister-item-content')
movies
```

```python
llist=soup.find_all('h3',{'class':'lister-item-header'} )
llist
```

```python
movie_title=[]
for x in llist:
    for y in x.find_all('a'):
        movie_title.append(y.text)
```

```python
director=[]
l_dir=soup.find_all('p',{'class':'text-muted text-small'} )
for j in l_dir[1: : 3]:
    for k in j.find('a'):
        director.append(k)
```

```python
income=[]
l_income=soup.find_all('p',{'class':'text-muted text-small'} )
for i in l_income[2: :3] :
    inc=i.text.split()[-1] 
    income.append(inc[1:-1])
```

```python
audience_rating=[]
aud_ratings=soup.find_all('div',{'class':'ipl-rating-star small'} )
for i in aud_ratings:
    a_rating=i.find_all('span')[1].text
    audience_rating.append(float(a_rating))
```

```python
critic_rating=[]
cri_ratings=soup.find_all('div',{'class':'inline-block ratings-metascore'} )
for i in cri_ratings:
    c_rating=i.text.split()[0]
    critic_rating.append(int(c_rating))
```

```python
data={"Movie_title": movie_title , "Director" : director ,"Income" : income ,
      "Audience Rating" :audience_rating , "Critic Rating" :critic_rating  }
```

```python
import pandas as pd
import numpy as np
lables=np.arange(45)

df=pd.DataFrame(data,index=lables)
df

```

```python
# sorting as per audience rating
df.sort_values(by="Audience Rating", ascending=False)
```

```python
# # sorting as per Critic rating
df.sort_values(by="Critic Rating", ascending=False)
```
